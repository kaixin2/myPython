1. Data Description

3.Data quality analysis

For data quality, there are two main problem categories. Firstly, there are some missing values. I use the function [.info()] to count the content of each column, and found that only four columns have no missing data, and the missing data of another ten columns is less than 100. Generally, there is a large amount of missing data in other columns.

Secondly, there is the problem of value inconsistency. Take column 'Did you calibrate your EC meter?' as an example. I used the function [.value_counts()] to make a statistical analysis of the value types in this column, and found that there are 11 kinds of values. But this column is the answer to the question. It should only be true or false, or yes or no. It should not be mixed or contain numbers. In addition, I use the function[.nunique()] to calculate several values in each column. There are many values in the columns of numerical records. For example, there are too many types of values of rainfall, which means that there is no standard for data processing (for example, divided into high, middle and low level).

Although the amount of data is defective, it is sufficient. Some data formats are not standard or have some errors, but the impact is not great. According to the NSW government data quality statement provided, the score of data quality is Institutional environment - 4, Accuracy - 4, Coherence - 5, Interpretability - 4, Accessibility - 3 (0-5, 5 is the best). In a word, the data quality is very good.

4.Exploratory analysis 

According to the volunteer handbook provided, the factors and rules for determining the water quality grades and scores provided are as the following table. I extracted the water quality data of each year and each summer, and analyzed it according to this table, and plotted the corresponding chart.

According to the notes to the table, the water quality grades of conductivity and salinity in the table are not applicable to estuarine ecosystems, but only to freshwater ecosystems. The data are taken from the Bellingerriver, so the conductivity and salinity are not analyzed in this report. In addition, there is no suspended solids data in the data provided, so this report does not analyze it.

For the PH value, Turbidity, Dissolved Oxygen and Faecal Coliforms, I make them into histograms to explore their distribution frequency. For the Available Phosphate value, I make it into box diagram to analyze the median and distribution range.

2017

The pH value in 2017 is mainly distributed in 6-8. According to the table, this means that water quality rating= good.
Turbidity in 2017 is mainly distributed in 7-10 NTU, which means water quality rating = good.
Available Phosphate in 2017 is mainly distributed in 0.02-0.08mg/L, among them, the values distributed in 0.05-0.08mg/L is more, which means water quality rating = good. There are a few extreme values in this group of data, but it does not affect the overall analysis. We eliminate the extreme values for analysis.
Dissolved Oxygen in 2017 distributed uneven, ranging from 10-105%, but mainly distributed in 70-100%, which means water quality rating = fair.
Faecal Coliforms in 2017 distributed mainly 0-100 CFU/100mL, which means water quality rating = good (Primary contact).
In a word, the overall water quality rating in 2017 is good.
